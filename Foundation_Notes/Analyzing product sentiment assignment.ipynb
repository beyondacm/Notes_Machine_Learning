```python
import graphlab
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

# Build a new feature with the counts for each of the selected words

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products
```

```python
'''
def awesome_count( w_dict ):
    if 'awesome' in w_dict :
        return w_dict['awesome']
    else :
        return 0
'''
def word_count(word, word_dict):
    if word in word_dict:
        return word_dict[word];
    else :
        return 0;
```

```python
products['awesome'] = products['word_count'].apply(lambda word_dict : word_count('awesome', word_dict))
```

```python
products
```

```python
products['awesome'].sum()
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda word_dict : word_count(word, word_dict))
```

```python
products
```

```python
for word in selected_words:
    print word
    print products[word].sum()
```

# Create a new sentiment analysis model using only the selected_words as features

```python
products = products[products['rating'] != 3]
# positive sentiment = 4 start or 5 start reviews
products['sentiment'] = products['rating'] >= 3
```

```python
train_data, test_data = products.random_split(.8, seed=0)
```

```python
train_data
```

```python
len(train_data)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data, 
                                                     target = 'sentiment',
                                                     features = selected_words,
                                                     validation_set = test_data)
```

```python
selected_words_model['coefficients'].sort('value')
```

# Comparing the accuracy of different sentiment analysis model: using the method 

```python
selected_words_model.evaluate(test_data)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data, 
                                                     target = 'sentiment',
                                                     features = ['word_count'],
                                                     validation_set = test_data)
```

```python
selected_words_model.show(view='Evaluation')
```

```python
sentiment_model.evaluate(test_data)
```
